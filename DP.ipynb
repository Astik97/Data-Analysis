```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot
```

```python
X,y = make_classification(n_samples=9000,n_features=12,n_informative=2,n_redundant=10,random_state=9)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
train_scores,test_scores=list(),list()
```

```python
values=[i for i in range(1,29)]
```

```python
for i in values:
    model=DecisionTreeClassifier(max_depth=i)
    model.fit(X_train,y_train)
    train_yhat=model.predict(X_train)
    train_acc=accuracy_score(y_train,train_yhat)
    test_yhat=model.predict(X_test)
    test_acc=accuracy_score(y_test,test_yhat)
    train_scores.append(train_acc)
    test_scores.append(test_acc)
    print('>%d,train:%.3f,test:%.3f'%(i,train_acc,test_acc))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
values=[i for i in range(10,29)]
```

```python
for i in values:
    model=DecisionTreeClassifier(max_depth=i)
    model.fit(X_train,y_train)
    train_yhat=model.predict(X_train)
    train_acc=accuracy_score(y_train,train_yhat)
    test_yhat=model.predict(X_test)
    test_acc=accuracy_score(y_test,test_yhat)
    train_scores.append(train_acc)
    test_scores.append(test_acc)
    print('>%d,train:%.2f,test:%.2f'%(i,train_acc,test_acc))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'criterion':['gini','entropy'],'max_depth':[2,4,6,8,10],'min_samples_split':[5,10,20,35,40]}
clf=GridSearchCV(DecisionTreeClassifier(),param_grid,cv=6,n_jobs=1,scoring="accuracy")
clf.fit(X_train,y_train)
```

```python
clf.best_estimator_
```

```python
print(accuracy_score(y_train,clf.best_estimator_.predict(X_train)))
print(accuracy_score(y_test,clf.best_estimator_.predict(X_test)))
```
