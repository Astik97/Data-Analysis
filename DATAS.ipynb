```python
import seaborn as sns
ax=sns.barplot(x='2002',data=df_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

This bar graph shows the value as per the year 2002.

```python
evaluate=df_data.groupby(['Country Name'],as_index=False)['Indicator Code'].sum().sort_values(by='Indicator Code',ascending=False).head(15)
sns.set(rc={'figure.figsize':(45,15)})
sns.barplot(data = evaluate,x='Country Name',y='Indicator Code')
```

This graph shows the dataset on different countries of the world on the basis of top 10 least developing countries
as per country name and indicator code.

```python
evaluate=df_data.groupby(['2010'],as_index=False)['2018'].sum().sort_values(by='2018',ascending=False).head(15)
sns.set(rc={'figure.figsize':(50,10)})
sns.barplot(data = evaluate,x='2010',y='2018')
```

This graph shows top 15 comparison of the year 2010 and 2018 dataset on different countries of the world and at the last value which increases rapidly.

```python
evaluate=df_data.groupby(['1990'],as_index=False)['1999'].sum().sort_values(by='1999',ascending=False).head(15)
sns.set(rc={'figure.figsize':(50,10)})
sns.barplot(data = evaluate,x='1990',y='1999')
```

This graph shows top 15 comparison of the year 2010 and 2018 dataset on different countries of the world,it increases at the third graph value.

```python
evaluate=df_data.groupby(['2000'],as_index=False)['2009'].sum().sort_values(by='2009',ascending=False).head(15)
sns.set(rc={'figure.figsize':(50,10)})
sns.barplot(data = evaluate,x='2000',y='2009')
```

This graph shows top 15 comparison of the year 2000 & 2009 dataset on different countries of the world which increases gradually 
but after second value its value increases normally.

```python
evaluate=df_data.groupby(['1990'],as_index=False)['2018'].sum().sort_values(by='2018',ascending=False).head(15)
sns.set(rc={'figure.figsize':(50,10)})
sns.barplot(data = evaluate,x='1990',y='2018')
```

This graph shows top 15 comparison of the year 1990 and 2018 dataset on different countries of the world which increases first and then decreases
after of it increases normally.
